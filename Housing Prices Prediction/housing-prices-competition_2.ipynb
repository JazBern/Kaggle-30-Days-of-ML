```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# Load the data, and separate the target
file_path = '../input/home-data-for-ml-course/train.csv'
home_data = pd.read_csv(file_path)
#Separating the target column
train_y = home_data.SalePrice
```

```python
# Finding the features most related to the target value

corr = home_data.corr() # df is the pandas dataframe
c1 = corr.abs()
c2 = c1['SalePrice']
c2.sort_values()[-19:-1]

```

```python
features = ['OpenPorchSF','2ndFlrSF','WoodDeckSF','LotFrontage','BsmtFinSF1','Fireplaces','MasVnrArea','GarageYrBlt','YearRemodAdd','YearBuilt','TotRmsAbvGrd','FullBath', '1stFlrSF','TotalBsmtSF','GarageArea','GarageCars','GrLivArea','OverallQual']
```

```python
train_data = home_data[features]
# Replacing NaN and Infinite values to 0
train_data = train_data.replace((np.inf, -np.inf), np.nan)
train_data.describe()
```

```python
from sklearn.impute import SimpleImputer

# Fill in the lines below: imputation
my_imputer = SimpleImputer()
imputed_train = pd.DataFrame(my_imputer.fit_transform(train_data))

# Fill in the lines below: imputation removed column names; put them back
imputed_train.columns = train_data.columns
```

```python
np.any(np.isnan(imputed_train))
```

```python
np.all(np.isfinite(imputed_train))
```

```python
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(imputed_train,train_y)
```

```python
# path to file you will use for predictions
test_data = '../input/home-data-for-ml-course/test.csv'

# read test data file using pandas
test = pd.read_csv(test_data)

# create test_X which comes from test_data but includes only the columns you used for prediction.
# The list of columns is stored in a variable called features
test_X = test[features]
test_X = test_X.replace((np.inf, -np.inf), np.nan)
```

```python
from sklearn.impute import SimpleImputer

# Fill in the lines below: imputation
my_imputer = SimpleImputer()
imputed_test = pd.DataFrame(my_imputer.fit_transform(test_X))

# Fill in the lines below: imputation removed column names; put them back
imputed_test.columns = test_X.columns
```

```python
# make predictions which we will submit. 
test_preds = rf_model.predict(imputed_test)
```

```python
# Run the code to save predictions in the format used for competition scoring

output = pd.DataFrame({'Id': test.Id,
                       'SalePrice': test_preds})
output.to_csv('submission.csv', index=False)
```
